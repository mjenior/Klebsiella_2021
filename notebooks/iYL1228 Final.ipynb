### Dictionary 'for' loop

```python
sequences = ['SRR8603247', 'SRR8603248']
```

```python
 for seq in sequences:
    locals()["dictionary" + str(seq)] = {}
    counter = 0 

    with open(str(seq) + '.mapped.tsv', 'r') as reads:
        for line in reads:
            line = line.split()
            gene_len = float(line[1])
            hits = float(line[2])
            if hits != 0: 
                normalized_numb = (hits * 150) / gene_len
                seq_name = line[0]
                seq_name = 'KPN_' + seq_name[-5:]
                locals()["dictionary"+str(seq)].update( {seq_name : normalized_numb} )
                counter = counter + 1 
```
